### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
fact_t1 = tienda['Precio'].sum()
fact_t2 = tienda2['Precio'].sum()
fact_t3 = tienda3['Precio'].sum()
fact_t4 = tienda4['Precio'].sum()

print("Facturación Tienda 1:", fact_t1)
print("Facturación Tienda 2:", fact_t2)
print("Facturación Tienda 3:", fact_t3)
print("Facturación Tienda 4:", fact_t4)
```

```python
import matplotlib.pyplot as plt

facturacion = {
    "Tienda 1": fact_t1,
    "Tienda 2": fact_t2,
    "Tienda 3": fact_t3,
    "Tienda 4": fact_t4
}

plt.figure()
plt.bar(facturacion.keys(), facturacion.values())
plt.title("Facturación Total por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Facturación Total")
plt.show()
```

El gráfico de facturación permite comparar el desempeño financiero de las cuatro tiendas. Se observa que la Tienda 1 presenta los ingresos más altos, lo que indica un mayor volumen de ventas o productos de mayor valor. Por otro lado, la Tienda 4
registra la facturación más baja, lo que podría reflejar menor demanda, menor ticket promedio o menor competitividad en el mercado. Este indicador es clave para evaluar la rentabilidad general de cada tienda.

# 3. Calificación promedio de la tienda


```python
prom_t1 = tienda['Calificación'].mean()
prom_t2 = tienda2['Calificación'].mean()
prom_t3 = tienda3['Calificación'].mean()
prom_t4 = tienda4['Calificación'].mean()

print("Promedio Tienda 1:", round(prom_t1,2))
print("Promedio Tienda 2:", round(prom_t2,2))
print("Promedio Tienda 3:", round(prom_t3,2))
print("Promedio Tienda 4:", round(prom_t4,2))
```

```python
calificaciones = [prom_t1, prom_t2, prom_t3, prom_t4]
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

plt.figure()
plt.plot(tiendas, calificaciones, marker='o')
plt.title("Promedio de Calificación por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Calificación Promedio")
plt.ylim(0,5)
plt.show()
```

El gráfico muestra el promedio de calificación de los clientes por tienda, lo cual refleja el nivel de satisfacción del consumidor. La Tienda 3 tiene la mejor valoración promedio, lo que sugiere una experiencia positiva de compra. En contraste, la Tienda 1 presenta la calificación más baja, lo que podría indicar oportunidades de mejora en calidad del producto, servicio o logística. La satisfacción del cliente es un factor determinante en la fidelización y el crecimiento del negocio.

# 5. Envío promedio por tienda

```python
envio_t1 = tienda['Costo de envío'].mean()
envio_t2 = tienda2['Costo de envío'].mean()
envio_t3 = tienda3['Costo de envío'].mean()
envio_t4 = tienda4['Costo de envío'].mean()

print("Envío promedio Tienda 1:", round(envio_t1,2))
print("Envío promedio Tienda 2:", round(envio_t2,2))
print("Envío promedio Tienda 3:", round(envio_t3,2))
print("Envío promedio Tienda 4:", round(envio_t4,2))
```

```python
envios = [envio_t1, envio_t2, envio_t3, envio_t4]

plt.figure()
plt.bar(tiendas, envios)
plt.title("Costo de Envío Promedio por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Costo Promedio de Envío")
plt.show()
```

El gráfico de costos de envío promedio muestra diferencias entre las tiendas. La Tienda 1 tiene el costo de envío más elevado, lo que podría afectar negativamente la decisión de compra, considerando que el cliente asume este gasto. Por otro lado, la Tienda 4 presenta un costo promedio más bajo, lo que puede representar una ventaja competitiva y favorecer mayores ventas.
